```python
%matplotlib inline

import nivapy3 as nivapy
import pandas as pd
import matplotlib.pyplot as plt
```

# Map Norwegian ICPW stations

```python
xl_path = r'../../../Thematic_Trends_Report_2020/land_cover_request/icpw_land_cover_by_country_2020-02-11.xlsx'
df = pd.read_excel(xl_path, sheet_name='all_stn_locs')
df = df.query('country == "Norway"')
df.head()
```

```python
nivapy.spatial.quickmap(df,
                        popup='station_code',
                        aerial_imagery=True,
                        kartverket=True,
                        )
```
